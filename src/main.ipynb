```python
import sys
import os   
import glob
import dotenv
# import openai
dotenv.load_dotenv()
from Web.scrape import DataScraper
from Web.keyword import KeywordFinder
```

```python
scraper = DataScraper()

prompt_list = [
    "Student Visa Interview",
    "US Visa Interview",
    "Study Abroad Interview",
    "Visa Interview Preparation",
    "Student Visa Application Interview",
    "Student Visa Regulations", 
    "F-1 Visa Requirements", 
    "International Student Visa Laws", 
    "Student Immigration Policies", 
    "F-1 Visa Application Process", 
    "Student Visa Compliance", 
    "Visa Status for International Students", 
    "F-1 Visa Renewal Procedures", 
    "Immigration Law for Student Visas", 
    "Student Visa Eligibility Criteria", 
    "Student Visa Regulations in the United States",
    "F-1 Visa Requirements for International Students",
    "International Student Visa Laws and Regulations",
    "Policies for Student Immigration in the United States",
    "F-1 Visa Application Process Step-by-Step",
    "Student Visa Compliance Guidelines",
    "Visa Status for International Students in the U.S.",
    "Procedures for F-1 Visa Renewal",
    "Overview of Immigration Law for Student Visas",
    "Eligibility Criteria for Student Visas"
]

finder = KeywordFinder(prompt_list)

result = []

# Iterate over each prompt and process it
for prompt in prompt_list:
    for item in finder.find_similar_keywords(prompt):
        web_article = scraper.process(item, top_article_index_param = 100)
        result.extend(web_article)
```

```python
import pandas as pd
df = pd.DataFrame(result) 
df.to_csv('web_articles.csv', index=False)
```
